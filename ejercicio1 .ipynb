```python
fig,ax=plt.subplots(figsize=(100,10))
ejeX=np.arange(len(fechas))
ax.set_xticks(ejeX)
ax.set_xticklabels(fechas, rotation=90)
valoresX={}
for region in regiones:
    valoresX[region]=[]
    
for fecha in fechas:
    for region in regiones:
        valoresX[region].append(datos[fecha][region])
        
for region in regiones:
    ax.plot(valoresX[region],label=region)
ax.legend()
ax.set_ylabel("Catidad de casos")
ax.set_xlabel("Dias")
ax.set_title("Cantiudad de casos por dia en cada region de chile")
```
